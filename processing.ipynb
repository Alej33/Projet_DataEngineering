```python
import pandas as pd

df = pd.read_csv("file_name.csv")
df
```

```python
import re

#On enl√®ve les √©mojis
def remove_emoji(text):
    emoji_pattern = re.compile("["
                        u"\U0001F600-\U0001F64F"  
                        u"\U0001F300-\U0001F5FF"  
                        u"\U0001F680-\U0001F6FF"  
                        u"\U0001F1E0-\U0001F1FF"  
                        u"\U0001F1F2-\U0001F1F4"  
                        u"\U0001F1E6-\U0001F1FF"  
                        u"\U0001F600-\U0001F64F"
                        u"\U00002702-\U000027B0"
                        u"\U000024C2-\U0001F251"
                        u"\U0001f926-\U0001f937"
                        u"\U0001F1F2"
                        u"\U0001F1F4"
                        u"\U0001F620"
                        u"\u200d"
                        u"\u2640-\u2642"
                        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['tweets'] = df['tweets'].apply(remove_emoji)

df
```

```python
import re

keywords1 = ['trafic', 'probl√®me', 'panne', 'incident', 'ralentissement', 'retard', 'bagage', 'malaise']
keywords2 = ['r√©tabli', 'normale', 'normal', 'r√©gulier', 'progressivement', 'fin', 'reprise','retour']

# Expression r√©guli√®re pour rechercher les mots cl√©s
regex1 = re.compile('|'.join(keywords1), flags=re.IGNORECASE)
regex2 = re.compile('|'.join(keywords2), flags=re.IGNORECASE)
regex3 = r'https://\S+'
regex4 = r'#\w+'

# Filtrage des tweets contenant les mots cl√©s
df_problemes = df[df['tweets'].str.contains(regex1)]   #On garde seulement les tweets qui concerne un probl√®me sur la ligne
df_problemes = df[~df['tweets'].str.contains(regex2)]  #On enl√®ve les tweets qui concerne un r√©tablissement de la ligne
df_problemes['tweets'] = df['tweets'].str.replace(regex3, '', regex=True)  #On enl√®ve les liens twitter
df_problemes['tweets'] = df['tweets'].str.replace(regex4, '', regex=True)  #On enl√®ve les hashtags

df_problemes
```

```python
reasons = []

# On d√©finit une expr√©ssion r√©guli√®re pour extaire les raisons d'un probl√®me sur la ligne
pattern1 = re.compile(r'(raison|cause|origine|suite|r√©percussion)\s*:?((?!de|du|des|√†|pour|sur)[^\.]+)(?:[\.\s]|$)')

# On d√©finit une expr√©ssion r√©guli√®re pour extaire les expressions entre parenth√®se et entre crochet
pattern2 = re.compile(r'[(\[]([^()\[\]]+)[)\]]')

# On cr√©e une nouvelle colonne 'raison'
df_problemes['raison'] = ''

# On boucle √† chaque ligne pour extraire les raisons gr√¢ce √† l'expression r√©guli√®re
for index, row in df_problemes.iterrows():
    tweet = row['tweets']
    match = pattern1.search(tweet)
    if match:
        reason = match.group(2).strip()
        reason = re.sub(r'\b(de|du|des|√†)\b', '', reason, flags=re.IGNORECASE) #On enl√®ve les pr√©positions
        reason = re.sub(r'[\.,]\s.*', '', reason) #On coupe la phrase √† partir d'une virgule ou un point
    else:
        match = pattern2.search(tweet) #Si la raison n'a pas √©t√© trouv√©e avec le 1er patterne, on l'extrait avec le 2√®me patterne
        if match:
            reason = match.group(1).strip()
        else:
            reason = ''
    #On ajoute rien si aucune raison a √©t√© trouv√©e
    reasons.append(reason)

df_problemes['raison'] = reasons

df_problemes
```

```python
masque1 = df_problemes['raison'].str.contains('\d+\/\d+')
masque2 = df_problemes['raison'].str.contains('Port du masque')

df_problemes.loc[masque1, 'raison'] = ''  #On enl√®ve les valeurs sous la forme 5/6 par exemple dans la colonne 'raison'
df_problemes.loc[masque2, 'raison'] = ''  #On enl√®ve les valeurs 'Port du masque' dans la colonne raison
```

```python
df_problemes = df_problemes[df_problemes['raison'] != '']  #On garde seuelement les lignes o√π il y a une r√©elle raison

df_problemes
```

```python
from nltk import FreqDist, word_tokenize
from nltk.collocations import BigramAssocMeasures, BigramCollocationFinder
from spacy.lang.fr.stop_words import STOP_WORDS as fr_stop

reasons = list(df_problemes['raison'])

stopwords = list(fr_stop) + ["d'un","d'une", ',', '‚Äô', 'd', 'l', ')','(', 'train', 'station', 'trafic', 'perturb√©', '*', "'"]

all_reasons = ' '.join(reasons)
tokens = word_tokenize(all_reasons)
tokens_without_stopwords = [word for word in tokens if word.lower() not in stopwords]
freq = FreqDist(tokens_without_stopwords)

#Extraction des raisons les plus fr√©quentes
most_common_reasons = [word for word, count in freq.most_common(100)]

#On va cr√©er une colonne 'raison_keyword' pour avoir seulement le mot-cl√© pour chaque ligne
new_reasons = []

for reason in reasons:
    found = False
    for common_reason in most_common_reasons:
        if common_reason in reason:
            new_reasons.append(common_reason)
            found = True
            break
    if not found:
        new_reasons.append('')

df_problemes['raison_keyword'] = new_reasons

#On transforme les mots-cl√©s les plus fr√©quents afin d'avoir une meilleur compr√©hension dans les graphiques
dict_mapping = {
    'oubli√©': 'Bagage oubli√©',
    'Bagage': 'Bagage oubli√©',
    'Bagages': 'Bagage oubli√©',
    'bagage': 'Bagage oubli√©',
    'voyageur': 'Malaise voyageur',
    'incident': 'Incident technique/voyageur',
    "d'exploitation": "Incident d'exploitation",
    'panne': 'Panne signalisation/train',
    'malaise': 'Malaise voyageur',
    'Malaise': 'Malaise voyageur',
    'technique': 'Incident technique',
    'signalisation': 'Panne de signalisation',
    'difficult√©s': "Incident d'exploitation",
    'voies': 'Personne sur les voies',
    'voie': 'Personne sur les voies',
    'personnes': 'Personne sur les voies',
    'Personnes': 'Personne sur les voies',
    'signal': "Signal d'alarme",
    "d'alarme": "Signal d'alarme",
    "l'actionnement": "Signal d'alarme",
    'incidents': 'Incident technique/voyageur',
    'Mouvement': 'Mouvement social',
    's√©curit√©': 'Mesure de s√©curit√©'
}

df_problemes['raison_keyword'] = df_problemes['raison_keyword'].replace(dict_mapping)
df_problemes
```

```python
from elasticsearch import Elasticsearch

LOCAL = False

es_client = Elasticsearch(hosts=["localhost" if LOCAL else "http://localhost:9200"])
```

```python
from elasticsearch.helpers import bulk

index_name = "ratp_tweets"
mapping = {
    "mappings": {
        "properties": {
            "account": {"type": "keyword"},
            "tweet": {"type": "text"},
            "reasons": {"type": "keyword"},
            "reasons_keyword": {"type": "keyword"},
            "timestamp": {"type": "date"}
        }
    }
}

if es_client.indices.exists(index_name):
    es_client.indices.delete(index=index_name)
    
#On cr√©e l'index avec le mapping
es_client.indices.create(index=index_name, body=mapping)


documents = []

for index, row in df_problemes.iterrows():
    document = {
        "_index": index_name,
        "_source": {
            "account": row["Lignes"],
            "tweet": row["tweets"],
            "reasons": row["raison"],
            "reasons_keyword": row["raison_keyword"],
            "timestamp": pd.to_datetime(row["date"] + " " + row["heure"], format="%d-%m-%Y %H:%M:%S")
        }
    }
    documents.append(document)
    
bulk(es_client, documents)
```
