```python
import snscrape.modules.twitter as twitter
import pandas
import pandas as pd


content = []
date = []
usernames = []

#Lignes de m√©tros
for i in range(1, 15):
    username = 'Ligne{}_RATP'.format(i)
    print(username)
    tweets = twitter.TwitterUserScraper(username).get_items()

    
    limit = 2000
    count = 0
    for tweet in tweets:
        if tweet.rawContent.startswith("@") == True:  #On enl√®ve les r√©ponses, donc les tweets comments par @
            continue
        if count == limit:
            break
        content.append(tweet.rawContent)
        date.append(tweet.date)
        usernames.append(tweet.user.username)
        count += 1

#Lignes de RER    
Lignes_RER = ['RER_A', 'RERB', 'RERC_SNCF', 'RERD_SNCF']

for i in Lignes_RER:
    
    print(i)
    tweets = twitter.TwitterUserScraper(i).get_items()
    limit = 2000
    count = 0
    for tweet in tweets:
        if tweet.rawContent.startswith("@") == True:
            continue
        if count == limit:
            break
        content.append(tweet.rawContent)
        date.append(tweet.date)
        usernames.append(tweet.user.username)
        count += 1

Lignes_Tram = ['T1_RATP', 'T2_RATP', 'T3a_RATP', 'T3b_RATP', 'RERE_T4_SNCF', 'T5_RATP', 'T6_RATP', 'T7_RATP', 'T8_RATP','T9_IDFM', 'T11Express','T13_IDFM']

for i in Lignes_Tram:
    
    print(i)
    tweets = twitter.TwitterUserScraper(i).get_items()
    limit = 500
    count = 0
    for tweet in tweets:
        if tweet.rawContent.startswith("@") == True:
            continue
        if count == limit:
            break
        content.append(tweet.rawContent)
        date.append(tweet.date)
        usernames.append(tweet.user.username)
        count += 1

df = pd.DataFrame({'Lignes': usernames, 'tweets': content, 'date': date})

df
```

```python
df = pd.DataFrame({'Lignes': usernames, 'tweets': content, 'date': date})

df['date_complete'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S+%f')

# On extrait la colonne de date et d'heure dans deux colonnes s√©par√©es
df['date'] = df['date_complete'].dt.strftime('%d-%m-%Y')
df['heure'] = df['date_complete'].dt.time
df = df.drop(columns=['date_complete'])
df
```

```python
df['tweets'] = df['tweets'].replace('\n', '', regex=True)
df
```

```python
df.to_csv('file_name.csv', encoding='utf-8', index=False)
```
